```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 1: Loading the dataset 
df = pd.read_csv(r'C:\Users\USER\Downloads\Desktop\DataScience\imdb_top_1000.csv')
```

```python
# Step 2: Cleaning and preprocessing the data

# Removing unnecessary columns
columns_to_drop = ['Poster_Link', 'Released_Year', 'Certificate', 'Runtime', 'Overview', 'Meta_score',
                   'Star1', 'Star2', 'Star3', 'Star4', 'No_of_Votes', 'Gross']
df = df.drop(columns_to_drop, axis=1)

# Checking for missing values
df.isnull().sum()
```

```python
df.head()
#df['Genre']= df['Genre'].astype(str)
#df.dtypes
```

```python
# Step 3: Exploring the data and identifying the most popular genre

# Split the genres multi-valued string content into a list 
df['Genre'] = df['Genre'].apply(lambda x: [genre.strip() for genre in x.split(',')])

# Creating a new DataFrame with exploded genres
df_exploded = df.explode('Genre')
```

```python
# Step 4: Creating a new DataFrame containing the count of movies/TV shows in each genre
genre_counts = df_exploded['Genre'].value_counts().reset_index()
genre_counts.columns = ['Genre', 'Count']
```

```python
# Step 5: Visualizing the data using a bar chart
plt.bar(genre_counts['Genre'], genre_counts['Count'])
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Most Popular Genre of Movies/TV Shows')
plt.xticks(rotation=90)
plt.show()
```

```python
# Step 6: Identifying the director with the most top-rated movies/TV shows
director_counts = df['Director'].value_counts().reset_index()
director_counts.columns = ['Director', 'Count']
```

```python
# Step 7: Visualizing the data using a bar chart
plt.bar(director_counts['Director'].head(15), director_counts['Count'].head(15))
plt.xlabel('Director')
plt.ylabel('Count')
plt.title('Top Directors with Most Top-rated Movies/TV Shows')
plt.xticks(rotation=90)
plt.show()
```
